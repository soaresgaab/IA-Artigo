# **--> Divisão Treino x Teste**

```python
from sklearn.model_selection import StratifiedKFold
```

```python
X = df.iloc[:, 1:]
y = df["isFraud"]
```

```python
k = 5
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
```

```python
fold = 1
for train_index, test_index in skf.split(X, y):
    print(f"\nFold {fold}:")

    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    print("Treino - X:")
    print(X_train.head())
    print("Treino - y:")
    print(y_train.head())

    print("Teste - X:")
    print(X_test.head())
    print("Teste - y:")
    print(y_test.head())

    fold += 1
```

# **--> Modelagem: Regressão Logística**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
model = LogisticRegression(random_state=42, max_iter=2000, solver="newton-cholesky")
```

```python
model.fit(X_train, y_train)
y_LR_pred = model.predict(X_test)
print("Relatório de Classificação:\n", classification_report(y_test, y_LR_pred))
```

# **--> Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42, oob_score=True)
rf.fit(X_train, y_train)
```

```python
y_RF_pred = rf.predict(X_test)
```

```python
print("Relatório de Classificação:\n", classification_report(y_test, y_RF_pred))
```

# **--> Arvore de Decisão**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from collections import Counter
```

```python
model = DecisionTreeClassifier(criterion="gini", max_depth=2, random_state=42)
model.fit(X_train, y_train)
y_pred_tree_2 = model.predict(X_test)
print(classification_report(y_test, y_pred_tree_2))
```

# **--> Arvore de Decisão com Votação**

```python
trees = [
    DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42),
    DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=42),
    DecisionTreeClassifier(criterion="log_loss", max_depth=7, random_state=42),
]

for i, tree in enumerate(trees):
    tree.fit(X_train, y_train)
    print(f"Árvore {i+1} treinada.")
    y_pred_tree_2 = tree.predict(X_test)
    print(f"Relatório de Classificação - Árvore {i+1}")
    print(classification_report(y_test, y_pred_tree_2))
    print("-" * 60)
```

```python
def majority_vote(trees, X):
    predictions = [tree.predict(X) for tree in trees]
    votes = list(zip(*predictions))
    majority_predictions = [Counter(vote).most_common(1)[0][0] for vote in votes]
    return majority_predictions

y_pred_vote_2 = majority_vote(trees, X_test)

print(classification_report(y_test, y_pred_vote_2))
```
