## 3.1.1 - Scaler

##Aplicando o tratamento mencionado no artigo

Scale features using statistics that are robust to outliers.

This Scaler removes the median and scales the data according to the quantile range (defaults to IQR: Interquartile Range). The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile).

Centering and scaling happen independently on each feature by computing the relevant statistics on the samples in the training set. Median and interquartile range are then stored to be used on later data using the transform method.

Standardization of a dataset is a common preprocessing for many machine learning estimators. Typically this is done by removing the mean and scaling to unit variance. However, outliers can often influence the sample mean / variance in a negative way. In such cases, using the median and the interquartile range often give better results.

```python
from sklearn.preprocessing import RobustScaler
X = pd.DataFrame(df_2['Time'])
transformer = RobustScaler().fit(X)
transformer
df2_time = transformer.transform(X)
df2_time
```

```python
X = pd.DataFrame(df_2['Amount'])
transformer = RobustScaler().fit(X)
transformer
df2_amount = transformer.transform(X)
df2_amount
```

```python
df_2['Time'] = df2_time
df_2['Amount'] = df2_amount
df_2
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
X = df_2.iloc[:, :-1]
y = df_2.iloc[:, -1]

k = 5
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)


fold = 1
for train_index, test_index in skf.split(X, y):
    print(f"\nFold {fold}:")


    X_train_2, X_test_2 = X.iloc[train_index], X.iloc[test_index]
    y_train_2, y_test_2 = y.iloc[train_index], y.iloc[test_index]

    print("Treino - X:")
    print(X_train_2.head())
    print("Treino - y:")
    print(y_train_2.head())

    print("Teste - X:")
    print(X_test_2.head())
    print("Teste - y:")
    print(y_test_2.head())

    fold += 1
```

## 5.4 Decision tree voting

```python
trees = [
    DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42),
    DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=42),
    DecisionTreeClassifier(criterion="log_loss", max_depth=7, random_state=42),
]

for i, tree in enumerate(trees):
    tree.fit(X_train_2, y_train_2)
    print(f"Árvore {i+1} treinada.")
    y_pred_tree_2 = tree.predict(X_test_2)
    print(f"Relatório de Classificação - Árvore {i+1}")
    print(classification_report(y_test_2, y_pred_tree_2))
    print("-" * 60)
```

```python
def majority_vote(trees, X):
    predictions = [tree.predict(X) for tree in trees]
    votes = list(zip(*predictions))
    majority_predictions = [Counter(vote).most_common(1)[0][0] for vote in votes]
    return majority_predictions

y_pred_vote_2 = majority_vote(trees, X_test_2)

print(classification_report(y_test_2, y_pred_vote_2))
```
